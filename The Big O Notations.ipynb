#***How do we test the efficiency of our code?***



* When working with a system we need to consider how efficient our program code is. There are two measures when testing our code, we have ***Time and Space Complexity***

* ***What is Time and Space complexity?***. These is how we measure the efficiency of our code. With time complexity, we want to measure how fast the runtime is for our code as input ***n*** gets relatively large.

***Consider the code below:***

```python
import timeit

# Function 1
def function1():
    # Some computation
    result = sum(range(1000000))
    return result

# Function 2
def function2():
    # Some different computation
    result = 0
    for i in range(1000000):
        result += i
    return result

# Measure the runtime of function 1
timeFunction1 = timeit.timeit("function1()", setup="from __main__ import function1", number=100)

# Measure the runtime of function 2
timeFunction2 = timeit.timeit("function2()", setup="from __main__ import function2", number=100)

print("Runtime of Function 1:", timeFunction1)
print("Runtime of Function 2:", timeFunction2)

if timeFunction1 < timeFunction2:
    print("Function 1 is faster.")
else:
    print("Function 2 is faster.")
```

* To explain the code above, we basically have two functions named func1 and func2. The goal is to find ***which of the two functions is faster?*** with the help of ***timeit*** import. So try ***pressing the play button located right before the codeblock and see what happens***

* According to the output, func1 is faster. It scores a time of 2.05s unlike func 2 which has a score of almost 8s.

* So the question now is ***Why is func1 faster than the other?*** To answer that question, you need to take note that when testing the time complexity of a program, you ***do not want to measure the*** ***exact runtime of the program***. This is actually due to various reasons but commonly its because of ***differences between the computer itself***.

* The reason why func1 is faster is because of the use of the ***sum() function***. ***This function is available in Python and it is optimized to take all the elements inside a list and add all the elements together unlike func2 which uses for loop to add all the elements together***.

* We could also measure algorithms by using the ***Big O notation*** and it could be written as ***0(n) (read as Big O of N)***.

# ***How do we determine the Big O notation of a program?***

***Observe the Code below:***


```python
#O(1) Constant

array = [1, 2, 3, 4, 5]

first_element = array[0]
print(f"The first element is: {first_element}")
```

* ***What do you think the Big O notation of the code above?***

* To explain the code above, we basically have an array which contains five elements. We then assign the first element of the array to the ***first_element variable***. We then print the first element of the array.

* ***The code example above is an example of an O(1) or Constant***. The reason why its O(1) or Constant ***is because no matter how many elements or numbers are inside the array, the program will still print the first element of the array which is 1.***


***Observe the code below:***

```python
#O(n) Linear

def take_list(lst):
  for number in lst:
    print(f"Element inside is: {number}")

array = [0,1,2,3,4,5]
take_list(array)
```

* Ill cut you off some slack. ***The code above is an example of a O(n) or Linear. The reason for that is simple, it is because the runtime of this code depends on how many elements or values are inside the array.***

* Remember the func1 and func2 functions above? ***They are both examples of O(n) or Linear!***

***Consider the code below:***

```python
#O(n^2) Quadratic

def quadratic_sum(array):
    total_sum = 0
    n = len(array)
    for i in range(n):  # O(n)
        for j in range(n):  # O(n)
            total_sum += array[i] + array[j]  # O(1)
    return total_sum

#Create Array and print result
arr = [1, 2, 3, 4, 5]
result = quadratic_sum(arr)
print("Result:", result)
```

* ***The code above is an example of a O(n^2) or Quadratic*** because for element ***i*** it iterates over the array again to find pairs and its sum.
